```python
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=['the glass of milk','the glass of juice','the cup of tea','i am a good boy','i am a good developer','understand the meaning of words','your videos are good']
```

```python
sent
```

```python
voc_size=10000

onehot_rep=[one_hot(words,voc_size)for words in sent]
onehot_rep
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
# padding (make sentence of same length)
# pre --> zeros at starting
sent_length=8
embedded_docs=pad_sequences(onehot_rep,padding='pre',maxlen=sent_length)
embedded_docs
```

```python
dim=10 # feature representation one word is rep by 10 
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
